```python
import seaborn as sns
import matplotlib.pyplot as plt

# rnek veri
data = sns.load_dataset("flights") 
data = data.pivot("month", "year", "passengers")

# Heatmap
sns.heatmap(data=data, annot=True, fmt="d", cmap="rocket_r")

plt.show()
```

```python
data = sns.load_dataset("flights")

# Line plot
sns.lineplot(data=data, x="year", y="passengers", hue="month")
```

```python
data = sns.load_dataset("iris")

# Swarm plot

sns.swarmplot(x="species", y="petal_length", data=data) # spiiz
```

```python
data = sns.load_dataset("titanic")

# Bar plot
sns.countplot(data=data, x="class")

plt.show()
```

```python
data = sns.load_dataset("penguins")
sns.displot(data=data, x="flipper_length_mm", kde=True)
```

```python
sns.displot(data=data, x="flipper_length_mm", hue="species", multiple="stack")
```

```python
data = sns.load_dataset("titanic")
sns.violinplot(data=data, x="class", y="age", hue="alive")
```

```python
data = sns.load_dataset("penguins")

sns.pairplot(data, hue="species", markers=["o", "s", "D"])
```
