# Matplotlib

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
a=np.arange(-50,50,1)
```

```python
a
```

```python
n=np.arange(0,100,1)
```

```python
m=np.sqrt(n)
```

```python
c=np.power(a,2)
```

```python
d=np.power(a,3)
```

```python
plt.plot(a,c,c="y")
plt.xlabel("numbers")
plt.ylabel("power")
plt.title("power 2 graph")
```

```python
plt.plot(a,d,c="g")
plt.xlabel("numbers")
plt.ylabel("power")
plt.title("power 3 graph")
```

```python
plt.plot(n,m,c="red")
plt.xlabel("numbers")
plt.ylabel("square")
plt.title("sqrt graph")
```

```python
x = np.random.rand(100)
y= np.random.rand(100)
colors = np.random.rand(100)
sizes = np.random.randint(100,200,100)

plt.scatter(x,y, c=colors, s=sizes)
plt.title("Pint Chart")
plt.xlabel("X axis")
plt.ylabel("Y axis")
```

```python

data = np.random.randn(1000)
plt.hist(data, bins=40,color="r")

# Determine chart properties
plt.title("Histogram")
plt.xlabel("Values")
plt.ylabel("Frequency")

# Show graphics
plt.show()
```

```python
sizes = [30, 44, 15, 10, 5, 5]

plt.pie(sizes)

plt.title("Pie Chart")
plt.show()
```

```python

x = ['A', 'B', 'C', 'D', 'E']
y = np.random.randint(1, 10, 5)
plt.bar(x, y)
plt.title("Column Chart")
plt.xlabel("Categories")
plt.ylabel("Values")
plt.show()
```
